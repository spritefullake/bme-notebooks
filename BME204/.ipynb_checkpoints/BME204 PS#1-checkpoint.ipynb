 Calculate the reaction energy (per mole) for the polymerization of:
 
 
 Reaction energy = **Energy released from making bonds (exothermic)** minus **Energy input to break the bonds (endothermic)** 
 
The system is the surroundings. 
* So when I *break* bonds, from my standpoint I am LOSING energy (even though the atoms are GAINING energy)
* When I *make* bonds, from my standpoint I am GAINING energy (energy is released from bond formation to the surrounding)

```python
# Break
cc_double = 680 # kJ/mol
# Form 
cc_single = 740 # kJ/mol
rxn_energy = lambda e_make, e_break : e_make - e_break
assert rxn_energy(cc_single, cc_double) == 60
```

Calculate the molecular weight of a molecule with n=700

```python
from math import *
# Rounds to the correct number of sig figs
def sig(x, n=3): 
    return round(x, -int(floor(log10(x))) + (n - 1))

n = 700 # molecules
N = 6.022E23 # molecules/mol
c_am = 12.011 # g/mol
f_am = 18.998 # g/mol
mol_molar_mass = 2*c_am + 4*f_am # g/mol (or amu)
answer = mol_molar_mass * n # grams per mole also describes amus!
assert sig(answer, 3) == 70000 # g
```
